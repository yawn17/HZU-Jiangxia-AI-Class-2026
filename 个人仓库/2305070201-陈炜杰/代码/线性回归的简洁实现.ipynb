1.生成数据集

```python
import numpy as np
import torch
from torch.utils import data
from d2l import torch as d2l
true_w = torch.tensor([2,-3.4])
true_b = 4.2
features, labels = d2l.synthetic_data(true_w, true_b, 1000)
```

2.读取数据集

```python
def load_array(data_arrays, batch_size, is_train=True): #@save
    " ""构造一个PyTorch数据迭代器"""
    dataset = data.TensorDataset(*data_arrays)
    return data.DataLoader(dataset, batch_size, shuffle=is_train)

batch_size = 10
data_iter = load_array((features, labels), batch_size)

next(iter(data_iter))
```

data.TensorDataset(*data_arrays) 会将输入的特征和标签打包成一个 TensorDataset，这样每次取出一个样本时，会同时返回一对（特征，标签）

接着，data.DataLoader 会根据指定的 batch_size 和 shuffle（是否打乱数据）参数，生成一个可以迭代的小批量数据加载器

代码设置了 batch_size = 10，然后调用 load_array((features, labels), batch_size) 得到 data_iter。这样，data_iter 就是一个可以在训练循环中使用的迭代器，每次迭代会返回一个包含 10 个样本的小批量数据。这样做的好处是可以充分利用硬件加速，提高训练效率，并且通过打乱数据（shuffle）可以提升模型的泛化能力。

3.模型的定义

```python
from torch import nn
net = nn.Sequential(nn.Linear(2, 1))
```

4.初始化模型

```python
net[0].weight.data.normal_(0, 0.01)
net[0].bias.data.fill_(0)
```

5.定义损失函数

```python
loss = nn.MSELoss()
```

6.定义优化算法

```python
trainer = torch.optim.SGD(net.parameters(), lr=0.03)
```

7.训练

```python
num_epochs = 3
for epoch in range(num_epochs):
    for X, y in data_iter:
        l = loss(net(X) ,y)
        trainer.zero_grad()
        l.backward()
        trainer.step()
        trainer.zero_grad()
    l = loss(net(features), labels)
    print(f'epoch {epoch + 1}, loss {l:f}')
```
