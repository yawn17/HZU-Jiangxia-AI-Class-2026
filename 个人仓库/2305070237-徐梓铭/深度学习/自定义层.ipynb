```python
import torch
import torch.nn.functional as F
from torch import nn
```

```python
class CenterredLayer(nn.Module):
    def __init__(self,):
        super().__init__()
    def forward(self,x):
        return x-x.mean()
layer=CenterredLayer()
layer(torch.FloatTensor([1,2,3,4,5]))
```

```python
net=nn.Sequential(nn.Linear(8,128),CenterredLayer())
y=net(torch.rand(4,8))
y.mean()
```

```python
class MyLinear(nn.Module):
    def __init__(self, in_units, units):
        super().__init__()
        self.weight = nn.Parameter(torch.randn(in_units, units))  # 修正: in_units
        self.bias = nn.Parameter(torch.randn(units,))
    
    def forward(self, X):
        linear = torch.matmul(X, self.weight.data) + self.bias.data  # 修正: weight
        return F.relu(linear)

dense = MyLinear(5, 3)
dense.weight 
```

```python
dense(torch.rand(2,5))
```

```python
net=nn.Sequential(MyLinear(64,8),MyLinear(8,1))
net(torch.rand(2,64))
```
