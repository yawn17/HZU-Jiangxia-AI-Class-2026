```python
# 课程类，保存课程信息
class Course:
    def __init__(self, name, price):
        self.name = name  # 课程名称
        self.price = price  # 课程价格
        
    def show(self):
        return f"{self.name}课程，价格{self.price}元"
```

```python
# 班级类，一个班级对应一个课程
class Class:
    def __init__(self, name, course):
        self.name = name  # 班级名称
        self.course = course  # 班级对应的课程
        self.students = []  # 这个班的学生列表
        self.teacher = None  # 这个班的老师
    
    # 添加学生到班级
    def add_student(self, student):
        self.students.append(student)
        student.my_class = self
        
    # 设置班级的老师
    def set_teacher(self, teacher):
        self.teacher = teacher
        
    # 显示班级信息
    def show_info(self):
        teacher_name = self.teacher.name if self.teacher else "暂无老师"
        return f"{self.name}，课程：{self.course.name}，老师：{teacher_name}，学生人数：{len(self.students)}人"
```

```python
# 学生类
class Student:
    def __init__(self, stu_id, name, campus):
        self.stu_id = stu_id  # 学号
        self.name = name  # 姓名
        self.campus = campus  # 所在校区
        self.my_class = None  # 报名的班级
        self.has_paid = False  # 是否缴费
        
    # 报名班级
    def join_class(self, class_obj):
        if self.my_class:
            return f"{self.name}已经报了{self.my_class.name}了"
        
        self.my_class = class_obj
        class_obj.add_student(self)
        return f"{self.name}报名了{class_obj.name}"
    
    # 缴费
    def pay(self):
        if not self.my_class:
            return "还没有报名班级"
            
        if self.has_paid:
            return "已经缴过费了"
            
        self.has_paid = True
        self.campus.money += self.my_class.course.price
        return f"{self.name}缴费{self.my_class.course.price}元"
    
    # 转校区
    def change_campus(self, new_campus):
        # 从原校区移除
        if self in self.campus.students:
            self.campus.students.remove(self)
        
        # 添加到新校区
        self.campus = new_campus
        new_campus.students.append(self)
        return f"{self.name}从{self.campus.name}转到{new_campus.name}"
    
    # 退学
    def quit(self):
        if not self.my_class:
            return "没有报名班级，不用退学"
            
        # 从班级中移除
        self.my_class.students.remove(self)
        # 从校区移除
        self.campus.students.remove(self)
        self.my_class = None
        
        return f"{self.name}退学了"
    
    # 显示学生信息
    def show_info(self):
        class_name = self.my_class.name if self.my_class else "无"
        paid = "已缴费" if self.has_paid else "未缴费"
        return f"学生{self.name}，学号{self.stu_id}，在{self.campus.name}，班级{class_name}，{paid}"
```

```python
# 老师类
class Teacher:
    def __init__(self, tea_id, name, campus):
        self.tea_id = tea_id  # 工号
        self.name = name  # 姓名
        self.campus = campus  # 所在校区
        self.classes = []  # 教的班级
        
    # 添加班级
    def add_class(self, class_obj):
        self.classes.append(class_obj)
        class_obj.set_teacher(self)
        
    # 显示老师信息
    def show_info(self):
        class_names = [c.name for c in self.classes]
        return f"老师{self.name}，工号{self.tea_id}，在{self.campus.name}，教{len(self.classes)}个班：{class_names}"
```

```python
# 员工类
class Staff:
    def __init__(self, staff_id, name, campus):
        self.staff_id = staff_id  # 工号
        self.name = name  # 姓名
        self.campus = campus  # 所在校区
        
    def show_info(self):
        return f"员工{self.name}，工号{self.staff_id}，在{self.campus.name}"
```

```python
# 校区类
class Campus:
    def __init__(self, name):
        self.name = name  # 校区名称
        self.money = 0  # 账户余额
        self.teachers = []  # 校区老师
        self.staffs = []  # 校区员工
        self.students = []  # 校区学生
        
    # 添加老师
    def add_teacher(self, teacher):
        self.teachers.append(teacher)
        
    # 添加员工
    def add_staff(self, staff):
        self.staffs.append(staff)
        
    # 添加学生
    def add_student(self, student):
        self.students.append(student)
        
    # 显示校区信息
    def show_info(self):
        return f"{self.name}：余额{self.money}元，老师{len(self.teachers)}人，员工{len(self.staffs)}人，学生{len(self.students)}人"
```

```python
# 总部类
class Headquarters:
    def __init__(self):
        self.campuses = []  # 所有校区
        
    # 添加校区
    def add_campus(self, campus):
        self.campuses.append(campus)
        
    # 显示所有校区信息
    def show_all(self):
        print("所有校区信息：")
        for campus in self.campuses:
            print(campus.show_info())
        
    # 统计总数
    def show_total(self):
        total_money = 0
        total_teachers = 0
        total_staffs = 0
        total_students = 0
        
        for campus in self.campuses:
            total_money += campus.money
            total_teachers += len(campus.teachers)
            total_staffs += len(campus.staffs)
            total_students += len(campus.students)
            
        print(f"总计：{total_money}元，{total_teachers}个老师，{total_staffs}个员工，{total_students}个学生")
```

```python
# 主程序开始

# 1. 创建总部
head = Headquarters()

# 2. 创建两个校区
campus1 = Campus("北京校区")
campus2 = Campus("上海校区")
head.add_campus(campus1)
head.add_campus(campus2)

print("创建了两个校区")

# 3. 创建课程
python_course = Course("Python编程", 5800)
java_course = Course("Java开发", 6200)
web_course = Course("前端开发", 5500)

print("创建了三个课程")

# 4. 创建班级
class1 = Class("Python一班", python_course)
class2 = Class("Java一班", java_course)
class3 = Class("Web一班", web_course)

# 5. 创建老师
teacher1 = Teacher("T001", "张老师", campus1)
teacher2 = Teacher("T002", "李老师", campus2)
teacher3 = Teacher("T003", "王老师", campus1)

campus1.add_teacher(teacher1)
campus2.add_teacher(teacher2)
campus1.add_teacher(teacher3)

# 分配老师教课
teacher1.add_class(class1)
teacher2.add_class(class2)
teacher3.add_class(class3)

print("创建了老师并分配了班级")

# 6. 创建员工
staff1 = Staff("S001", "刘行政", campus1)
staff2 = Staff("S002", "赵行政", campus2)

campus1.add_staff(staff1)
campus2.add_staff(staff2)

# 7. 创建学生
student1 = Student("2024001", "小明", campus1)
student2 = Student("2024002", "小红", campus1)
student3 = Student("2024003", "小刚", campus2)
student4 = Student("2024004", "小丽", campus2)

campus1.add_student(student1)
campus1.add_student(student2)
campus2.add_student(student3)
campus2.add_student(student4)

print("创建了学生")

# 8. 学生报名班级
print("\n学生报名")
print(student1.join_class(class1))
print(student2.join_class(class1))
print(student3.join_class(class2))
print(student4.join_class(class3))

# 9. 学生缴费
print("\n学生缴费")
print(student1.pay())
print(student2.pay())
print(student3.pay())
print(student4.pay())

# 10. 显示信息
print("\n学生信息")
print(student1.show_info())
print(student2.show_info())
print(student3.show_info())
print(student4.show_info())

print("\n老师信息")
print(teacher1.show_info())
print(teacher2.show_info())
print(teacher3.show_info())

print("\n员工信息")
print(staff1.show_info())
print(staff2.show_info())

print("\n班级信息")
print(class1.show_info())
print(class2.show_info())
print(class3.show_info())

# 11. 校区信息
print("\n校区信息")
head.show_all()

# 12. 测试转校
print("\n测试转校")
print(student2.change_campus(campus2))
head.show_all()

# 13. 测试退学
print("\n测试退学")
print(student4.quit())
head.show_all()

# 14. 总计
print("\n总部统计")
head.show_total()
```
