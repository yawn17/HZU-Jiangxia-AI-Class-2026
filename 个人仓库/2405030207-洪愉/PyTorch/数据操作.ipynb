```python
X = torch.arange(12,dtype=torch.float32).reshape((3,4))
Y = torch.tensor([[2.0,1,4,3],[1,2,3,4],[4,3,2,1]])
torch.cat((X,Y),dim=0),torch.cat((X,Y),dim=1) # 拼接张量，dim=0 按行拼接，dim=1 按列拼接
```

```python
X == Y
```

```python
X.sum()
```

```python
# 创建两个不同形状的张量，用于广播机制演示
a = torch.arange(3).reshape((3, 1))
b = torch.arange(2).reshape((1, 2))
a, b
```

```python
a + b
```

```python
# 原地操作：创建与Y形状相同的零张量，然后修改其值，id不变
Z = torch.zeros_like(Y)  
print('id(Z):', id(Z))  
Z[:] = X + Y         
print('id(Z):', id(Z))
```

```python
before = id (X)
X += Y
id(X) == before
```

```python
# 张量与NumPy数组的相互转换
A = X.numpy()
B = torch.tensor(A)
type(A),type(B)
```
