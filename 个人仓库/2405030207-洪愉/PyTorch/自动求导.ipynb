```python
import torch
```

```python
x = torch.arange(4.0)
x
```

```python
x.requires_grad_(True)  #等价于'x =torch.arange(4.0,requires_grad=True)'
x.grad #默认值是None
```

```python
y = 2*torch.dot(x,x) 
y
```

```python
y.backward() # 反向传播求梯度
x.grad
```

```python
x.grad == 4 * x # 验证梯度是否正确
```

```python
# 在默认情况下，PyTorch会累积梯度，我们需要清除之前的值
x.grad.zero_()
y = x.sum()
y.backward()
x.grad
```

```python
# 再次清零，对 y = x*x 逐元素求导
x.grad.zero_()
y=x*x
y.sum().backward()
x.grad
```

```python
# 使用 detach 分离计算图
x.grad.zero_()
y = x*x
u = y.detach() # u 视为常数，不再记录梯度
z = u*x

z.sum().backward()
x.grad == u
```

```python
# 验证 y 的梯度
x.grad.zero_()
y.sum().backward()
x.grad == 2 * x
```

```python
# python控制流下的自动求导
def f(a):
    b = a*2
    while b.norm() <1000:
        b = b*2
    if b.sum()>0:
        c = b
    else:
        c = 100*b
    return c

a = torch.randn(size=(), requires_grad=True)
d = f(a)
d.backward()

# 验证梯度
a.grad == d / a
```
