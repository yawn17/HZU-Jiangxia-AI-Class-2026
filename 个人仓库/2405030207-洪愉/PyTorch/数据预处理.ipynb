```python
import os

data_dir = r'C:\Users\hoyu\Desktop\data'

# 创建目录
os.makedirs(data_dir, exist_ok=True)

# 构建CSV文件的完整绝对路径
data_file = os.path.join(data_dir, 'house_tiny.csv')

# 写入数据到CSV文件
with open(data_file, 'w', encoding='utf-8') as f:
    f.write('NumRooms,Alley,Price\n')  # 列名
    f.write('NA,Pave,127500\n')        # 每行表示一个数据样本
    f.write('2,NA,106000\n')
    f.write('4,NA,178100\n')
    f.write('NA,NA,140000\n')

print(f"数据集已创建并保存到：{data_file}")
print("请到桌面查看 data 文件夹中的 house_tiny.csv 文件")
```

```python
import pandas as pd

data = pd.read_csv(data_file) 
print(data)
```

```python
# 分离输入特征和输出标签
inputs , outputs = data.iloc[:,0:2] , data.iloc[:,2]
# 用数值列的均值填充缺失值
inputs = inputs.fillna(inputs.mean(numeric_only=True))
print(inputs)
```

```python
inputs = pd.get_dummies(inputs, dummy_na=True)  # 将类别特征转换为独热编码，并处理缺失值作为单独一类
inputs = inputs.astype(int) # 将布尔值转换为整数
print(inputs)
```
