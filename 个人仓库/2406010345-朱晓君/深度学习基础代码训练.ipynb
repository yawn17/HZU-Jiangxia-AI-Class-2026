```python
#激活函数
```

```python
import torch
import torch.nn as nn
x=torch.tensor([-2.0,-1.0,0.0,1.0,2.0])
```

```python
#Sigmoid(输出0~1之间的数)
sigmoid=nn.Sigmoid()
print("Sigmoid:",sigmoid(x))
```

```python
#ReLU（负数变0）
relu=nn.ReLU()
print("ReLU:",relu(x))
```

```python
#GELU(高斯误差线性单元)
gelu=nn.GELU()
print("GELU:",gelu(x))
```

```python
#SiLU
silu=nn.SiLU()
print("SiLU:",silu(x))
```

```python
#进行一个可视化对比
import matplotlib.pyplot as plt
x_plot=torch.linspace(-3,3,100)
plt.plot(x_plot,nn.ReLU()(x_plot),label='ReLU')
plt.plot(x_plot,nn.Sigmoid()(x_plot),label='Sigmoid')
plt.plot(x_plot,nn.GELU()(x_plot),label='GELU')
plt.plot(x_plot,nn.SiLU()(x_plot),label='SiLU')
plt.legend();plt.grid(True);plt.title("激活函数对比");plt.show()
```

```python
#损失函数
```

```python
import torch
import torch.nn as nn
```

```python
pred=torch.randn(4,5)
target=torch.tensor([0,2,1,4])
```

```python
#交叉熵损失 CrossEntropyLoss=Softmax+NLLLoss
```

```python
ce_loss=nn.CrossEntropyLoss()
loss1=ce_loss(pred,target)
print("交叉熵损失：",loss1.item())
```

```python
#KL散度损失
log_prob=nn.LogSoftmax(dim=1)(pred)
prob_target=nn.Softmax(dim=1)(torch.randn(4,5))
k1_loss=nn.KLDivLoss(reduction='batchmean')
loss2=k1_loss(log_prob,prob_target)
print("KL散度损失：",loss2.item())
```

```python
#MSELoss均方误差（回归任务）
mse_loss=nn.MSELoss()
pred_reg=torch.randn(4,1)
target_reg=torch.randn(4,1)
loss3=mse_loss(pred_reg,target_reg)
print("MSE损失：",loss3.item)
```

```python
#L1Loss平均绝对误差
l1_loss=nn.L1Loss()
loss4=l1_loss(pred_reg,target_reg)
print("L1损失：",loss4.item())
```

```python
import torch
import torch.nn as nn
```

```python
#归一化BatchNorm
bn=nn.BatchNorm2d(num_features=16)
x=torch.randn(4,16,32,32)
out=bn(x)
print(f"输入均值：{x.mean():.3f}，方差：{x.var():.3f}")
print(f"输出均值：{out.mean():.3f}，方差：{out.var():.3f}")
```

```python
#残差连接
class ResidualBlock(nn.Module):
    def __init__(self,dim):
        super().__init__()
        self.linear1=nn.Linear(dim,dim)
        self.linear2=nn.Linear(dim,dim)
        self.relu=nn.ReLU()
    def forward(self,x):
        residual=x
        out=self.linear1(x)
        out=self.relu(out)
        out=self.linear2(out)
        out=out+residual
        return self.relu(out)
res_block=ResidualBlock(64)
x=torch.randn(10,64)
out=res_block(x)
print(f"输入形状：{x.shape}，输出形状：{out.shape}")
print(f"输入输出关系：out=F(x)+x")
```

```python
#Dropout随机失活
dropout=nn.Dropout(p=0.5)
x=torch.ones(1,10)
out_train=dropout(x)
print("训练模式输出：",out_train)
dropout.eval()
out_eval=dropout(x)
print("评估模式输出:",out_eval)
print("\n训练模式下多次执行：")
dropout.train()
for i in range(5):
    out=dropout(x)
    print(f"第{i+1}次：{out.squeeze().tolist()}")
```
