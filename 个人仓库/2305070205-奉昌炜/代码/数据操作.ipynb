## 节省内存

```python
before = id(Y)  #将Y的内存地址存起来
Y = Y + X       #进行运算
id(Y) == before #与原来地址比较
```

##### 运行一些操作可能会导致为新结果分配内存，这是不可取的，所以我们需要原地更新内存地址

```python
Z = torch.zeros_like(Y)
print('id(Z):', id(Z))
Z[:] = X + Y
print('id(Z):', id(Z))
```

##### ^^^这是使用切片表示法将操作的结果分配给先前分配的数组

```python
before = id(X)
X += Y
id(X) == before
```

##### 我们可以使用X[:] = X + Y或X += Y来减少操作的内存开销
